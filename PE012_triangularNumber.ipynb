The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

     1: 1
     3: 1,3
     6: 1,2,3,6
    10: 1,2,5,10
    15: 1,3,5,15
    21: 1,3,7,21
    28: 1,2,4,7,14,28

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

```python
import math
```

```python
# determine the number of factors that n has. We only need to check up to ~sqrt(n).
def numFactors(n):
    maxFactor = math.ceil(math.sqrt(n)) - 1

    tot = 0
    for i in range(1,maxFactor+1):
        if n % i == 0:
            tot += 1

    # account for whether n is a perfect square
    if math.ceil(math.sqrt(n)) == math.sqrt(n):
        return 2*tot + 1
    else:
        return 2*tot
            
    return None
```

```python
# Realistically, the first number to have 501 divisors will be at least 251^2: this allows for all the numbers 
# 1-250 as divisors, along with 251 as well as a factor larger than 251 that pairs with each factor in 1-250.
minTriangleNum = 251**2;

triangleNum = 1
lastAdded = 1
factors = 1
while factors <= 500:
    lastAdded += 1
    triangleNum += lastAdded
    if triangleNum < minTriangleNum:
        continue
    
    factors = numFactors(triangleNum)
```
